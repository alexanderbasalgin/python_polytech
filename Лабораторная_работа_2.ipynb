### Задание №5*

Выведите список (в виде объекта list) уникальных авторов за 2011 год, чье количество отзывовов хотя бы на одну книгу было больше 75 квантиля.

```python
df_amazon_2011 = df_amazon[df_amazon['Year'] == 2011]
author_reviews = df_amazon_2011.groupby('Author')['Reviews'].sum()
quantile = author_reviews.quantile(0.75)
popular_authors = author_reviews[author_reviews > quantile]
answer_list = popular_authors.index.tolist()
answer_list
```

### Задание №6

```python
df_amazon['genre_categories'] = df_amazon['Genre'].astype('category')
df_amazon['genre_codes'] = df_amazon['genre_categories'].cat.codes
df_amazon
```
